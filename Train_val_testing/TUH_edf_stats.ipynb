```python
import pandas as pd
import os
import numpy as np

SAMPLING_RATE = 250
csv_path = '/home/gayal/ssl-project/PatchTST/preprocessing/inputs/sub_list2_2000.csv'
df = pd.read_csv(csv_path)

# plot the histogram for 'time_len' column
# diide the values by 250 before that
df['time_len_sec'] = df['time_len'] / SAMPLING_RATE
# plot the histogram
import matplotlib.pyplot as plt
plt.hist(df['time_len_sec'], bins=5)
plt.xlabel('Time Length (seconds)')
plt.ylabel('Frequency')
plt.title('Histogram of Time Lengths')
# plt.savefig('time_len_histogram.png')
plt.show()

# plot in minutes
df['time_len_min'] = df['time_len_sec'] / 60
# plot the histogram
plt.hist(df['time_len_min'], bins=5)
plt.xlabel('Time Length (minutes)')
plt.ylabel('Frequency')
plt.title('Histogram of Time Lengths')
# plt.savefig('time_len_histogram_min.png')
plt.show()

# plot in hours
df['time_len_hours'] = df['time_len_min'] / 60
# plot the histogram
plt.hist(df['time_len_hours'], bins=5)
plt.xlabel('Time Length (hours)')
plt.ylabel('Frequency')
plt.title('Histogram of Time Lengths')
plt.savefig('time_len_histogram_hours.png')
plt.show()
```
